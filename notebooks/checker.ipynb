````python
def get_dists(true_embeds, false_embeds, test_true_embeds, V_topic=None, use_V_topic=False, plot=True):
    import numpy as np
    from sklearn.metrics.pairwise import cosine_similarity

    if use_V_topic:
        if V_topic is None:
            V_topic = np.mean(true_embeds, axis=0).reshape(1, -1)

        true_dist = cosine_similarity(test_true_embeds, V_topic).reshape(-1)
        false_dist = cosine_similarity(false_embeds, V_topic).reshape(-1)
    else:
        true_dist = np.max(cosine_similarity(test_true_embeds, true_embeds), axis=1)
        false_dist = np.max(cosine_similarity(false_embeds, true_embeds), axis=1)

    if plot:
        import matplotlib.pyplot as plt
        import seaborn as sns

        sns.histplot(true_dist, color="blue", bins=20)
        sns.histplot(false_dist, color="red", bins=20)

    return true_dist, false_dist
````

````python
from sentence_transformers import SentenceTransformer

all_MiniLM_L6_v2 = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
````

````python
true_embeddings_miniLM = all_MiniLM_L6_v2.encode(true)
false_embeddings_miniLM = all_MiniLM_L6_v2.encode(false)
test_true_embeddings_miniLM = all_MiniLM_L6_v2.encode(test_true)
V_topic_miniLM = all_MiniLM_L6_v2.encode(topic)
````

````python
true_dist_miniLM_max, false_dist_miniLM_max = get_dists(true_embeddings_miniLM, 
                                                false_embeddings_miniLM, 
                                                test_true_embeddings_miniLM, 
                                                use_V_topic=False, plot=False)

true_dist_miniLM_mean, false_dist_miniLM_mean = get_dists(true_embeddings_miniLM, 
                                                false_embeddings_miniLM, 
                                                test_true_embeddings_miniLM, 
                                                use_V_topic=True, plot=False)

true_dist_miniLM_semantic, false_dist_miniLM_semantic = get_dists(true_embeddings_miniLM, 
                                                false_embeddings_miniLM, 
                                                test_true_embeddings_miniLM, 
                                                V_topic=V_topic_miniLM,
                                            use_V_topic=True, plot=False)

import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(true_dist_miniLM_max, color="blue", bins=20, alpha=0.6, label='True', ax=axes[0])
sns.histplot(false_dist_miniLM_max, color="red", bins=20, alpha=0.6, label='False', ax=axes[0])
axes[0].set_title('MAX')
axes[0].legend()

sns.histplot(true_dist_miniLM_mean, color="blue", bins=20, alpha=0.6, label='True', ax=axes[1])
sns.histplot(false_dist_miniLM_mean, color="red", bins=20, alpha=0.6, label='False', ax=axes[1])
axes[1].set_title('V_topic (mean)')
axes[1].legend()

sns.histplot(true_dist_miniLM_semantic, color="blue", bins=20, alpha=0.6, label='True', ax=axes[2])
sns.histplot(false_dist_miniLM_semantic, color="red", bins=20, alpha=0.6, label='False', ax=axes[2])
axes[2].set_title('V_topic (semantic cloud)')
axes[2].legend()

plt.tight_layout()
plt.show()
````
